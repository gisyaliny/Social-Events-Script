## Transfer Social Events to Shapefile

```python
import geopandas as gpd
import numpy as np
import pandas as pd
```

```python
import sys
sys.path.append('Script/data_mining/functions/')
import merge_df
```

```python
df = pd.read_excel('events.xlsx')
df = merge_df.clean_index(df)
df.head()
```

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Long, df.Lat))
gdf.crs = {'init': 'epsg:4326'}
gdf = gdf.to_crs(epsg=2163)
gdf.head()
```

```python
shp_name = 'Shapefile/Social-Events/social_events.shp'
gdf.to_file(filename= shp_name)
```

## Read the Zillow neighbors

```python
zillow_neighbors = 'Shapefile/Zillow/Dallas_Neighborhoods.shp'
zillow_gdf = gpd.read_file(zillow_neighbors)
print('Zillow_neighbors', zillow_gdf.crs)
print('Social-Events', gdf.crs)
```

```python
zillow_gdf = zillow_gdf.to_crs(epsg=2163)
print('Zillow_neighbors', zillow_gdf.crs)
print('Social-Events', gdf.crs)
```

### Take a look of datasets

```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.colors import ListedColormap
plt.rcParams['figure.figsize'] = [16, 12]
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
zillow_gdf.plot(cmap='Greys', ax=ax,alpha=.5)
gdf.plot(ax=ax)
plt.axis('equal')
ax.set_axis_off()
plt.show()
```

```python
# Clip the data using GeoPandas clip
neighbor_events = gpd.clip(gdf, zillow_gdf)
neighbor_events.head()
```

```python
fig, ax = plt.subplots(figsize=(18, 12))
zillow_gdf.plot(cmap='Greys',edgecolor="black",ax=ax,alpha=.5)

for ctype,data in neighbor_events.groupby('Category'):
    data.plot(label = ctype,
             ax = ax)
ax.set_axis_off()
plt.axis('equal')
plt.legend()
plt.show()
```

```python
shp_name = 'Shapefile/Social-Events/neighbor_social_events.shp'
neighbor_events.to_file(shp_name )
```

```python
group_events = neighbor_events.groupby(['Group','Category']).size().reset_index(name = 'Frequncy')
group_events.head(20)
```
